```python
import math
import matplotlib.pyplot as plt
import  matplotlib
import numpy as np

from scipy import linalg, stats
```

```python
def f(x):
    return math.sin(2*x) / math.cos(x)
```

# Solution 1

```python
a,b = 2,3
```

```python
step=(b-a)/10
values = []
points = []
while a <= b:
    points.append(a)
    values.append(f(a))
    a += step
print(len(values))
print(len(points))
```

# Solution 2

```python
plt.rc('font', size=20)
plt.rc('axes', labelsize=15)
plt.rc('legend', fontsize=10)
    
plt.title('My function (2 solution)',fontsize=28)
plt.plot(points,values)

plt.legend(['math.sin(2*x) / math.cos(x)'])
plt.ylabel('y(values)')
plt.xlabel('x(points)')

plt.show()
```

# Solution 3

```python
plt.style.use('default')
plt.plot(points, values, marker='p', markerfacecolor=[0.3, 0.1, 0.7, 1], markersize=10)
plt.title('Solution 3')
plt.grid(color=([0.3,0.2,1]), linestyle='-.', linewidth=0.6, alpha=1)

plt.ylabel('y(values)')
plt.xlabel('x(points)')

plt.show()
```

# Solution 6

```python
def func(x, y):
    return (x - 3)**2 + (y - 8)**2

x = np.random.random(5)
y = np.random.random(5)

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection='3d')
ax.plot(x,y, func(x,y), color=[1, 0, 0, 1])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x)')
ax.set_title(' ')

plt.show()
```
